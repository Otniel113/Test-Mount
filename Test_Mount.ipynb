# Read and Modify

```python
import pandas as pd
```

Jika format berupa spreadsheet (bukan .xlsx maupun .csv)

```python
sheet_id = '1bfnaznn2tAd9BwsA5-RruTWYtA6r7_1rA6AYq5MrScw'
```

```python
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"
```

```python
data_by_url = pd.read_csv(url)
data_by_url
```

Untuk format .xlsx atau .csv

```python
data_by_directory = pd.read_excel('Kehadiran.xlsx')
data_by_directory
```

Modifikasi, misalkan tambah row

```python
new_row = {'No': 4,
           'Nama': 'Chandra',
           'Keterangan':'Hadir',
          }

data_by_directory = pd.concat([data_by_directory, pd.DataFrame([new_row])], ignore_index=True)
data_by_directory
```

# Export

Membuat file excel baru di Drive yg sama dgn file sebelumnya. Jika ingin overwrite, bisa mengganti nama file menjadi sama dengan nama di awal

```python
data_by_directory.to_excel(r'Kehadiran2.xlsx', index=False, header=True)
```
